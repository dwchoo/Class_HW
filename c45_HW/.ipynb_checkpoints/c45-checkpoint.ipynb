```python
def train_decision_tree(name, train_set, test_set, criterion='entropy', max_depth = 2,verbose = 0):
    from sklearn.tree import DecisionTreeClassifier
    from sklearn.metrics import classification_report, confusion_matrix
    from sklearn.tree import export_graphviz
    from IPython.display import display
    import graphviz
    
    tree_clf = DecisionTreeClassifier(criterion=criterion, max_depth=max_depth)
    features = list(train_set.columns[:-1])
    X = train_set[features]
    y = train_set['target']
    
    tree_clf.fit(X,y)
    
    file_name = './graph_imag/wine_tree_{}'.format(name)
    file_name_dot = '{}.dot'.format(file_name)
    export_graphviz(
        tree_clf,
        out_file= file_name_dot,
        feature_names=features,
        rounded=True,
        filled=True
    )
    
    with open(file_name_dot) as f:
        dot_graph = f.read()
    dot = graphviz.Source(dot_graph)
    dot.format = 'png'
    dot.render(filename=file_name, directory='./', cleanup=True)
    
    
    
    X_test = test_set[features]
    y_test = test_set['target']
    y_pred = tree_clf.predict(X_test)
    cf_matrix = confusion_matrix(y_test,y_pred)
    report = classification_report(y_test,y_pred, output_dict = True)
    report_print = classification_report(y_test,y_pred)
    report['matrix'] = cf_matrix
    report['print'] = report_print
    report['criterion'] = criterion
    report['depth'] = max_depth
    
    if verbose == 1:
        print('Criterion : {}    max_depth : {}\n'.format(criterion,max_depth))
        display(dot)
        print(report['matrix'])
        print(report['print'])
    return report
```

```python
entropy_2 = train_decision_tree(
    name = 'entropy_2',
    train_set=train_set,
    test_set=test_set,
    criterion='entropy',
    max_depth=2,
    verbose=1
)
```

```python
entropy_list  = [entropy_2, entropy_3, entropy_4, entropy_5, entropy_6, entropy_7, entropy_8]
gini_list = [gini_2, gini_3, gini_4, gini_5, gini_6, gini_7, gini_8]
entropy_result_dict = {
    'precision' : [],
    'recall'    : [],
    'f1'        : [],
    'accuracy'  : [],
    'criterion' : [],
    'depth'     : []
}
gini_result_dict = {
    'precision' : [],
    'recall'    : [],
    'f1'        : [],
    'accuracy'  : [],
    'criterion' : [],
    'depth'     : []
}
total_result_dict = {
    'precision' : [],
    'recall'    : [],
    'f1'        : [],
    'accuracy'  : [],
    'criterion' : [],
    'depth'     : [],
}
for data in entropy_list:
    entropy_result_dict['precision'] += [data['macro avg']['precision']]
    entropy_result_dict['recall'   ] += [data['macro avg']['recall'   ]]
    entropy_result_dict['f1'       ] += [data['macro avg']['f1-score' ]]
    entropy_result_dict['accuracy' ] += [data['accuracy']              ]
    entropy_result_dict['criterion'] += [data['criterion']             ]
    entropy_result_dict['depth'    ] += [data['depth']                 ]
    
for data in gini_list:
    gini_result_dict['precision'] += [data['macro avg']['precision']]
    gini_result_dict['recall'   ] += [data['macro avg']['recall'   ]]
    gini_result_dict['f1'       ] += [data['macro avg']['f1-score' ]]
    gini_result_dict['accuracy' ] += [data['accuracy']              ]
    gini_result_dict['criterion'] += [data['criterion']             ]
    gini_result_dict['depth'    ] += [data['depth']                 ]
    
for data in entropy_list+gini_list:
    total_result_dict['precision'] += [data['macro avg']['precision']]
    total_result_dict['recall'   ] += [data['macro avg']['recall'   ]]
    total_result_dict['f1'       ] += [data['macro avg']['f1-score' ]]
    total_result_dict['accuracy' ] += [data['accuracy']              ]
    total_result_dict['criterion'] += [data['criterion']             ]
    total_result_dict['depth'    ] += [data['depth']                 ]

entropy_result_df = pd.DataFrame(entropy_result_dict)
gini_result_df    = pd.DataFrame(gini_result_dict)
total_result_df = pd.DataFrame(total_result_dict)
```

# 결론
---------
### 1. 문제 3번 - 지니 불순도(Gini)와 엔트로피(Entropy)로 바꾸었을 때 훈련된 결정 트리의 성능에 차이가 있는 지 설명하시오  
* 지니 불순도(Gini impurity)  
$ \displaystyle I_G(p) = \sum_{i=1}^j p_i \sum_{k \ne i}p_k = 1- \sum_{i = 1}^j {p_i}^2 $  

* 정보 엔트로피(Information entropy)  
$ \displaystyle H(T) = I_E(p_1,p_2,\cdots,p_J) = - \sum_{i-1}^J P_i \log_2 p_i  $  
  
Gini impurity와 Information entropy의 식은 위와 같다. 위에 보이는 식과 같이 두 식의 차이는 크지 않다. 실제로 두 식을 통해서 만들어지는 Decision tree도 극적인 차이를 보여주지는 않는다. 하지만 Entropy를 계산할 때 log를 계산해야 하는 단점은 많은 데이터를 다룰 때 퍼포먼스 하락을 야기하기도 한다. 그리고 이번 실험의 결과를 보았을 때 Decision tree의 max depth가 깊어질 수록 Gini impurity를 사용했을 때 성능이 더 좋음을 볼 수 있었다.  
근본적으로 Gini impurity와 Information entropy 모두 Decision tree의 노드를 구성하기 위해서 사용되는 지표이고 같은 항목에 대한 Decision 결과는 같은 결과를 보여주지만 tree의 depth가 깊어짐에 따라 서로 다른 항목을 선택하여 tree의 형태가 달라져 다른 결과를 보여주었다. 따라서 데이터의 개수가 많은 빅데이터를 다룰 경우 속도 면에서 우수한 Gini impurty를 사용하는 것도 좋은 선택지가 될 수 있다.
<br>

### 2. 문제4번 - max_depth의 바꾸었을 때 Tree의 모양과 성능에 어떠한 차이가 있는지 설명하시오.  
max\_depth는 Tree의 depth를 나타낸다. 즉 max\_depth가 2일 경우 tree의 depth는 2가 되고 3일 경우 depth가 3이 되는 것이다.

그리고 깊이가 깊어져도 더 이상 분류할 수 없는 경우 결정트리는 노드를 더 만들지 않고 멈춘다.  
Decision tree의 depth를 늘릴수록 성능이 좋아지는 경향을 보이지만 일정 수준 이상 넘어갈 경우 더 이상 효과가 없거나 성능이 하락하는 것을 결과 그래프를 통해서 볼 수 있었다. 이는 과접합(Overfitting) 현상으로 보인다. 옵션으로 주어진 depth가 깊어지면서 학습 데이터에 대해서 계속 노드를 늘려가면서 학습한다. 그리하여 학습 데이터에 대해서는 100%에 가까운 성능을 내지만 학습하지 않은 Test 데이터에 대해서는 잘 판단하지 못하는 모델이 만들어 진다.  
학습 데이터에 대해서도 잘 동작하고 test 데이터에 대해서도 잘 동작하는 모델을 만들기 위해서는 많은 상황을 가지고 실험을 해야 한다. 현재의 결과도 하나의 random sampling된 학습 데이터와 test 데이터에 대한 결과 이므로 특정 depth가 잘 동작하는 depth고 좋은 모델이라고 말할 수 없다. 한정된 데이터를 가지고 성능을 평가하는 N-fold cross validation 방법 등 다양한 방법을 통해서 최적 모델을 찾는 과정이 필요하다.  
결론적으로 일정 depth 까지는 성능이 좋아지지만 일정 수준부터는 효과가 없거나 overfitting으로 인해서 test 데이터애 대해 성능이 하락하는 결과를 보여준다.
