```python
def train_decision_tree(name, train_set, test_set, criterion='entropy', max_depth = 2,verbose = 0):
    from sklearn.tree import DecisionTreeClassifier
    from sklearn.metrics import classification_report, confusion_matrix
    from sklearn.tree import export_graphviz
    from IPython.display import display
    import graphviz
    
    tree_clf = DecisionTreeClassifier(criterion=criterion, max_depth=max_depth)
    features = list(train_set.columns[:-1])
    X = train_set[features]
    y = train_set['target']
    
    tree_clf.fit(X,y)
    
    file_name = './graph_imag/wine_tree_{}'.format(name)
    file_name_dot = '{}.dot'.format(file_name)
    export_graphviz(
        tree_clf,
        out_file= file_name_dot,
        feature_names=features,
        rounded=True,
        filled=True
    )
    
    with open(file_name_dot) as f:
        dot_graph = f.read()
    dot = graphviz.Source(dot_graph)
    dot.format = 'png'
    dot.render(filename=file_name, directory='./', cleanup=True)
    
    
    
    X_test = test_set[features]
    y_test = test_set['target']
    y_pred = tree_clf.predict(X_test)
    cf_matrix = confusion_matrix(y_test,y_pred)
    report = classification_report(y_test,y_pred, output_dict = True)
    report_print = classification_report(y_test,y_pred)
    report['matrix'] = cf_matrix
    report['print'] = report_print
    report['criterion'] = criterion
    report['depth'] = max_depth
    
    if verbose == 1:
        print('Criterion : {}    max_depth : {}\n'.format(criterion,max_depth))
        display(dot)
        print(report['matrix'])
        print(report['print'])
    return report
```

```python
entropy_2 = train_decision_tree(
    name = 'entropy_2',
    train_set=train_set,
    test_set=test_set,
    criterion='entropy',
    max_depth=2,
    verbose=1
)
```

## Result chart

```python
entropy_list  = [entropy_2, entropy_3, entropy_4, entropy_5, entropy_6, entropy_7, entropy_8]
gini_list = [gini_2, gini_3, gini_4, gini_5, gini_6, gini_7, gini_8]
entropy_result_dict = {
    'precision' : [],
    'recall'    : [],
    'f1'        : [],
    'accuracy'  : [],
    'criterion' : [],
    'depth'     : []
}
gini_result_dict = {
    'precision' : [],
    'recall'    : [],
    'f1'        : [],
    'accuracy'  : [],
    'criterion' : [],
    'depth'     : []
}
total_result_dict = {
    'precision' : [],
    'recall'    : [],
    'f1'        : [],
    'accuracy'  : [],
    'criterion' : [],
    'depth'     : [],
}
for data in entropy_list:
    entropy_result_dict['precision'] += [data['macro avg']['precision']]
    entropy_result_dict['recall'   ] += [data['macro avg']['recall'   ]]
    entropy_result_dict['f1'       ] += [data['macro avg']['f1-score' ]]
    entropy_result_dict['accuracy' ] += [data['accuracy']              ]
    entropy_result_dict['criterion'] += [data['criterion']             ]
    entropy_result_dict['depth'    ] += [data['depth']                 ]
    
for data in gini_list:
    gini_result_dict['precision'] += [data['macro avg']['precision']]
    gini_result_dict['recall'   ] += [data['macro avg']['recall'   ]]
    gini_result_dict['f1'       ] += [data['macro avg']['f1-score' ]]
    gini_result_dict['accuracy' ] += [data['accuracy']              ]
    gini_result_dict['criterion'] += [data['criterion']             ]
    gini_result_dict['depth'    ] += [data['depth']                 ]
    
for data in entropy_list+gini_list:
    total_result_dict['precision'] += [data['macro avg']['precision']]
    total_result_dict['recall'   ] += [data['macro avg']['recall'   ]]
    total_result_dict['f1'       ] += [data['macro avg']['f1-score' ]]
    total_result_dict['accuracy' ] += [data['accuracy']              ]
    total_result_dict['criterion'] += [data['criterion']             ]
    total_result_dict['depth'    ] += [data['depth']                 ]

entropy_result_df = pd.DataFrame(entropy_result_dict)
gini_result_df    = pd.DataFrame(gini_result_dict)
total_result_df = pd.DataFrame(total_result_dict)
```

### 1.  3 -  (Gini) (Entropy)            
*  (Gini impurity)  
$ \displaystyle I_G(p) = \sum_{i=1}^j p_i \sum_{k \ne i}p_k = 1- \sum_{i = 1}^j {p_i}^2 $  

*  (Information entropy)  
$ \displaystyle H(T) = I_E(p_1,p_2,\cdots,p_J) = - \sum_{i-1}^J P_i \log_2 p_i  $
